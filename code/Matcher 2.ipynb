```python
import pandas as pd


csv_file_path = "C:\\Users\\User\\Desktop\\Μεταπτυχιακό Business Analytcs ΟΠΑ\\3rd semester\\Machine Learning and Content Analysis\\styles.csv"

# Specify the column indices you want to read (0-based indexing)
columns_to_read = [0, 1, 3, 4, 5, 6, 8]

# Read the CSV file with the specified delimiter and columns
df = pd.read_csv(csv_file_path, sep=';', usecols=columns_to_read)


# Give an example vector with chothe characteristics
vec = ['Men', 'Topwear', 'Shirts', 'Navy Blue', 'Fall', 'Casual']


# Convert the vector to a DataFrame with the same column names (excluding 'id')
vec_df = pd.DataFrame([vec], columns=df.columns[1:])  # Exclude 'id'



# Filter the DataFrame based on the conditions
filtered_df = df[
    (df['season'] == vec_df['season'][0]) &
    (df['usage'] == vec_df['usage'][0]) &
    (
        ((df['subCategory'] == 'Bottomwear') & (vec_df['subCategory'][0] == 'Topwear')) |
        ((df['subCategory'] == 'Topwear') & (vec_df['subCategory'][0] == 'Bottomwear'))
    ) &
    (
        (df['gender'] == vec_df['gender'][0]) |
        (
            (df['gender'] == 'Unisex') &
            (vec_df['gender'][0] in ['Men', 'Women', 'Unisex'])
        ) |
        (
            (df['gender'] == 'Men') &
            (vec_df['gender'][0] in ['Men', 'Unisex'])
        ) |
        (
            (df['gender'] == 'Women') &
            (vec_df['gender'][0] in ['Women', 'Unisex'])
        ) |
        (
            (df['gender'] == 'Girls') &
            (vec_df['gender'][0] == 'Girls')
        ) |
        (
            (df['gender'] == 'Boys') &
            (vec_df['gender'][0] == 'Boys')
        )
    )
]


# Define color  rule
colors = [
    "blue", "black", "white", "grey", "green", "red", "navy blue",
    "purple", "pink", "yellow", "brown", "orange", "beige",
    "maroon", "cream", "olive", "multi", "charcoal"
]

color_compatibility_rules = {
    'Black': colors,
    'Blue': ['Black', 'White', 'Grey', 'Green', 'Red', 'Navy Blue', 'Purple', 'Pink'],
    'White': ['Black', 'Blue', 'Grey', 'Green', 'Red', 'Navy Blue', 'Purple', 'Pink'],
    'Grey': ['Black', 'Blue', 'White', 'Green', 'Red', 'Navy Blue', 'Purple', 'Pink'],
    'Green': ['Black', 'Grey', 'Navy Blue'],
    'Red': ['Black', 'Blue', 'White', 'Grey', 'Pink', 'Maroon', 'Multi', 'Charcoal'],
    'Navy Blue': ['Black', 'Blue', 'White', 'Grey', 'Green', 'Purple', 'Multi', 'Charcoal'],
    'Purple': ['Black', 'Blue', 'White', 'Grey', 'Navy Blue', 'Pink'],
    'Pink': ['Black', 'Blue', 'White', 'Grey', 'Red', 'Purple'],
    'Yellow': ['Black', 'Brown', 'Green', 'Olive'],
    'Brown': ['Black', 'Grey', 'Pink', 'Yellow', 'Olive'],
    'Orange': ['Black', 'Brown', 'Green', 'Olive'],
    'Beige': ['Black', 'Blue', 'Grey', 'Green', 'Pink'],
    'Maroon': ['Black', 'Grey', 'Red'],
    'Cream': ['Black', 'Blue', 'White', 'Grey', 'Green', 'Pink'],
    'Olive': ['Black', 'Brown', 'Green', 'Yellow', 'Orange'],
    'Multi': ['Black', 'Blue', 'White', 'Grey', 'Green', 'Red', 'Navy Blue', 'Purple'],
    'Charcoal': ['Black', 'Red', 'Navy Blue', 'Multi']
}


# Function to check color compatibility
def are_colors_compatible(color1, color2):
    return color2 in color_compatibility_rules.get(color1, [])


# Filter the dataset based on color compatibility
compatible_rows = []
for index, row in filtered_df.iterrows():
    if are_colors_compatible(vec[3], row['baseColour']):
        compatible_rows.append(row)
        
        
# Create a new DataFrame with the compatible rows
colored_data = pd.DataFrame(compatible_rows)

# Display the filtered dataset
print(colored_data)        


print('All OK')


```
